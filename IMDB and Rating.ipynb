```python
df = data[data['Demographic']=='All'][0:20]
df
```

```python
from pyecharts import Line

line = Line("2019 IMDb Top 20 Movies Ratings","AGE 18-29",width=800)
line.add("All Ages", df['title'], df['All Ages'], is_smooth=True, 
         legend_text_size=18,xaxis_label_textsize=14,yaxis_label_textsize=18,
         xaxis_rotate=100, yaxis_min=2)
line.add("18-29", df['title'], df['18-29'], is_smooth=True, 
         legend_text_size=18,xaxis_label_textsize=14,yaxis_label_textsize=18,
         xaxis_rotate=100, yaxis_min=2, mark_point=["max"])
line
```
